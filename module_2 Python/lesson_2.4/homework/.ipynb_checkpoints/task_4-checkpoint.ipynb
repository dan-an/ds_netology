## Задание 4

Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2', '5455 028765'],
    '2': ['10006', '5400 028765', '5455 002299'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):

p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;
s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится
Правильно обработайте ситуацию, когда пользователь будет вводить несуществующий документ.

```python
def get_doc_id(): 
    while True:
        doc_id = input("Введите номер документа: ")
        
        if doc_id == 'q':
            return
        elif any(document for document in documents if document.get('number') == doc_id):
            return doc_id
        else:
            show_warning('doc_not_found')
            continue
    
def get_name(doc_id):
    try:
        return next(document for document in documents if document['number'] == doc_id)['name']
    except:
        show_warning('doc_not_found')
        
def get_shelf(doc_id):
    
    for shelf, docs_list in directories.items():
        if doc_id in docs_list:
            return shelf
    
def get_list():
    docs_for_print = []
    for document in documents:
        docs_for_print.append(f"{document.get('type')} \"{document.get('number')}\" {document.get('name')}")
    
    return docs_for_print

def create_shelf():
    global directories
    
    while True:
        new_shelf_id = input("Введите номер новой полки: ")
        
        if new_shelf_id == 'q':
            return
        
        if new_shelf_id not in directories.keys():
            directories[new_shelf_id] = []
            print(f"Создана новая полка номер {new_shelf_id}")
            break
        else:
            show_warning('shelf_exists')
            continue
            
def delete_doc(doc_id):
    global documents
    global directories
    
    shelf_id = get_shelf(doc_id)
    
    documents = [document for document in documents if document.get('number') != doc_id]
    directories[shelf_id] = [document for document in documents if document != doc_id]
    print(f"Документ номер {doc_id} удален")
    print(documents)
    print(directories)
    
def move_doc(doc_id):
    target_shelf_id = None
    current_shelf_id = get_shelf(doc_id)
        
    while True:
        target_shelf_id = input("Введите номер новой полки: ")
        
        if target_shelf_id == 'q':
            return
        
        if target_shelf_id in directories.keys():
            directories[current_shelf_id] = [document for document in directories[current_shelf_id] if document != doc_id]
            directories[target_shelf_id].append(doc_id)
            print(f"Документ перемещен на полку {target_shelf_id}")
            break
        else:
            show_warning('shelf_not_found')
            continue
            
def add_doc():
    while True:
        new_doc_id = input("Введите номер документа: ")
        
        if new_doc_id == 'q':
            return
        
        if not any(document for document in documents if document.get('number') == new_doc_id):
            break
        else:
            show_warning('doc_exists')
            continue
            
    doc_type = input("Введите тип документа: ")
    if doc_type == 'q':
        return
    
    owner_name =  input("Введите имя владельца документа: ")
    if owner_name == 'q':
        return
        
    while True:
        target_shelf_id = input("Введите номер полки: ")
        
        if target_shelf_id == 'q':
            return
        
        if target_shelf_id in directories.keys():
            documents.append({'type': doc_type, 'number': new_doc_id, 'name': owner_name})
            directories[target_shelf_id].append(new_doc_id)
            print(f"Новый документ помещен на полку {target_shelf_id}")
            break
        else:
            show_warning('shelf_not_found')
            continue
    
        
def show_warning(error):
    errors = {
        'command_error': 'Несуществующая команда. Введите валидную команду. Для выхода введите q',
        'doc_not_found': 'Документ отсутствует в базе. Введите другой номер документа. Для выхода введите q',
        'doc_exists': 'Документ с таким номером уже существует. Введите другой номер нового документа. Для выхода введите q',
        'shelf_exists': 'Полка с таким номером уже существует. Введите другой номер новой полки. Для выхода введите q',
        'shelf_not_found': 'Полка с таким номером не найдена. Введите другой номер полки. Для выхода введите q'
    }
    
    print(errors[error])

    
def main():
    command_list = ['p', 's', 'l', 'as', 'd', 'm', 'a', 'q']
    command = None
    result = ''

    while command == None or command not in command_list:
        command = input("Введите команду: ")
        if command not in command_list: 
            show_warning('command_error')
        elif command == 'q': 
            return
        
    if command == 'p':
        doc_id = get_doc_id()
        if doc_id == None:
            return
        result = get_name(doc_id)
    elif command == 's':
        doc_id = get_doc_id()
        if doc_id == None:
            return
        result = get_shelf(doc_id)
    elif command == 'l':
        print(*get_list(), sep = '\n')
    elif command == 'as':
        create_shelf(),
    elif command == 'd':
        doc_id = get_doc_id()
        if doc_id == None:
            return
        delete_doc(doc_id),
    elif command == 'm':
        doc_id = get_doc_id()
        if doc_id == None:
            return
        move_doc(doc_id)
    elif command == 'a':
        add_doc()
        print(documents)
        print(directories)
        
    print(result if result != None else '')
    
    return
```

```python
main()
```
